```python
# Analysis 2 - Reporting  - who sent most business emails in the company
# Whose has done more business work

import glob
import csv
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from email.parser import Parser


business_terms=['market','gas','price','power','energy','trade','business','service']

def business_function(payload):
    word_list=payload
    tokens=tokenizer.tokenize(word_list)
    token_list=' '.join(word for word in tokens)
    return token_list
    

tokenizer = RegexpTokenizer(r'\w+')
count=0
business_email={}
for fname in glob.iglob("C:/Users/Nimi/midterm/data/Data/maildir/*/_sent_mail/*"):
    with open(fname,'r') as f:
        email=Parser().parsestr(f.read())
        token_list=business_function(email.get_payload())
        for x in business_terms:
            if x in token_list:
                from_email_id=email['From']
                if from_email_id not in business_email:
                    business_email[from_email_id]=1
                else:
                    business_email[from_email_id]=business_email[from_email_id]+1               
            break;

#Sort the dictionary based on value

sorted_business_emails=sorted(business_email.items(), key=lambda x: x[1])

#Put the Data in csv file

dir=os.getcwd()
with open(dir+"/"+'q1-analysis1', 'w',newline='') as csvfile:
    fieldnames = ['Email-id', 'No of Business Email sent']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    count=0;
    for key,value in sorted_business_emails:
        writer.writerow({'Email-id': key, 'No of Business Email sent': value})
```
